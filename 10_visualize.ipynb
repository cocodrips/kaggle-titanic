```python
from IPython.core.display import display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import patsy
import seaborn as sns

import base
import styles
base.show_basic_data()

import matplotlib.cm as cm
```

```python
display("Female")
pce_f_df:pd.DataFrame = base.titanic_female.groupby(['Pclass', 'Child', 'Embarked'])['Survived'].agg(['mean', 'count'])
display(pce_f_df.style.bar(subset=['count'], color='gray').applymap(styles.gradient_by_ratio, subset=['mean']))


display("Male")
pce_m_df:pd.DataFrame = base.titanic_male.groupby(['Pclass', 'Child', 'Embarked'])['Survived'].agg(['mean', 'count'])
display(pce_m_df.style.bar(subset=['count'], color='gray').applymap(styles.gradient_by_ratio, subset=['mean']))

```
